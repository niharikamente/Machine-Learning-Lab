```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler
from sklearn.impute import SimpleImputer
```

```python
data = pd.read_csv("train.csv")#load dataset
print(data.head()) #display the first five rows of dataset
```

```python
print(data.info())#no. of rows,cols,datatypes,missing values
print(data.describe())#statistical summary
print(data.isnull().sum())
```

```python
imputer = SimpleImputer(strategy='mean')
data[['Age']] = imputer.fit_transform(data[['Age']])
```

```python
data = data.drop_duplicates()
```

```python
le = LabelEncoder()
data['Sex'] = le.fit_transform(data['Sex'])
```

```python
data.head()
```

```python
print(data.duplicated())
```

```python
imputer = SimpleImputer(strategy="most_frequent")

data["Embarked"] = imputer.fit_transform(data[["Embarked"]]).ravel()
```

```python
print(data.columns)
```

```python
print(data.duplicated().sum())
```

```python
data["Embarked"] = le.fit_transform(data["Embarked"])
```

```python
data.head()
```

```python
scaler = MinMaxScaler()

data[["Age","Fare"]] = scaler.fit_transform(data[["Age","Fare"]])
```

```python
data.head()
```

```python
Q1 = data["Fare"].quantile(0.25)

Q3 = data["Fare"].quantile(0.75)

IQR = Q3 - Q1
```

```python
data = data[
    (data["Fare"] >= Q1 - 1.5 * IQR) &
    (data["Fare"] <= Q3 + 1.5 * IQR)
]
```

```python
data.to_csv("processed_train.csv", index=False)
```
