```python
import pandas as pd
import numpy as np

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
columns = [
    "Age","Workclass","fnlwgt","Education","EducationNum",
    "MaritalStatus","Occupation","Relationship","Race","Sex",
    "CapitalGain","CapitalLoss","HoursPerWeek","NativeCountry","Income"
]

data = pd.read_csv("adult.data", names=columns)

print(data.head())
```

```python
print(data.shape)
print(data.columns)
print(data.info())
```

```python
print(data.describe())
```

```python
print(data.isnull().sum())
```

```python
data.replace(" ?", np.nan, inplace=True)

print(data.isnull().sum())
```

```python
imputer = SimpleImputer(strategy="most_frequent")

categorical = ["Workclass","Occupation","NativeCountry"]

data[categorical] = imputer.fit_transform(data[categorical])
```

```python
print(data.duplicated().sum())
```

```python
data = data.drop_duplicates()
```

```python
le = LabelEncoder()

columns = [
    "Workclass","Education","MaritalStatus",
    "Occupation","Relationship","Race",
    "Sex","NativeCountry","Income"
]

for col in columns:
    data[col] = le.fit_transform(data[col])
```

```python
print(data.head())
```

```python
scaler = MinMaxScaler()

num_cols = [
    "Age","fnlwgt","EducationNum",
    "CapitalGain","CapitalLoss","HoursPerWeek"
]

data[num_cols] = scaler.fit_transform(data[num_cols])
```

```python
Q1 = data["Age"].quantile(0.25)
Q3 = data["Age"].quantile(0.75)

IQR = Q3 - Q1
```

```python
data = data[
    (data["Age"] >= Q1 - 1.5*IQR) &
    (data["Age"] <= Q3 + 1.5*IQR)
]
```

```python
# Save processed data
data.to_csv("adult_processed.csv", index=False)
```
